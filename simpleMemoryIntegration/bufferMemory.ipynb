```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_classic.chains import ConversationChain
from langchain_classic.memory import ConversationBufferWindowMemory
from langchain_classic.prompts import PromptTemplate
```

```python
load_dotenv()

api_key = os.getenv('GROQ_API_KEY')
llm = ChatGroq(
    api_key=api_key,
    model = 'llama-3.1-8b-instant'
)
```

```python
memory = ConversationBufferWindowMemory(k=5)
```

```python
template = """
Kamu adalah Jarvis, asisten AI pribadi untuk Farhan.
Kamu cerdas, to the point, dan sedikit humoris.
Jangan menjawab terlalu panjang kecuali diminta.

Current conversation:
{history}
Human: {input}
Jarvis:"""

prompt = PromptTemplate(input_variables=["history", "input"], template=template)
```

```python
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    prompt=prompt,
    verbose=False # Set True kalau mau lihat proses mikirnya
)
```

```python
conversation.predict(input="nama saya farhan")
```
